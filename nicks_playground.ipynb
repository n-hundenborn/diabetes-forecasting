```python
import pandas as pd
import os

df_list = []
for file in os.listdir('./data/'):
    if file[-3:] != "csv":
        continue

    print(f"Reading file: {file}")
    df_list.append(pd.read_csv(f"./data/{file}", index_col=False))
```

```python
rows = 0
cols = 0
for df in df_list:
    rows += df.shape[0]
    cols += df.shape[1]
print(rows)
print(cols)
df_all = pd.concat(df_list, ignore_index=True, join="inner")
df_all.shape
```
